```python
#seperating labels and target data
y=df['complication']

X=df.drop('complication',axis=1)
```

```python
#Splitting data into test and train data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
model=KNeighborsClassifier()
```

```python
rfc_params=[{'n_neighbors': list(np.arange(20))}]

clf=GridSearchCV(model,rfc_params,cv =5,scoring='accuracy',refit=True)
```

```python
#using GridSearchCV to get the best hyperparameters the  use the classifier with the paramaters
start_time=time.time()
clf.fit(X_train,y_train)
elapsed=time.time()-start_time
print('Time for training: '+str(elapsed))
print(clf.best_params_)
print(clf.best_index_)
print(clf.best_score_)
```

```python
model=KNeighborsClassifier(n_neighbors=9)
model.fit(X_train,y_train) 
y_pred=model.predict(X_test)
```

```python
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
